```python
import openpyxl as px
import numpy as np
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
import datetime
import pymysql.cursors
from dateutil.relativedelta import relativedelta
from collections import Counter
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"/System/Library/Fonts/STHeiti Light.ttc",size=12)
read_content = pd.read_excel('blue(19000篇).xlsx')
```

```python
db = pymysql.connect(host="localhost", user="root", password="", db="proj_blue_server",charset='utf8mb4',port = 3306)
cursor = db.cursor()
```

```python
#文章分類
cursor.execute("SELECT classify FROM llingjing_2")
classify = []
for i in cursor:
    classify.append(str(i).split("(")[1].split(",")[0].split("'")[1])
```

```python
#關鍵字
cursor.execute("SELECT notes1_1,notes1_2,notes1_3,notes2_1,notes3_1,notes4_1,notes5_1,notes6_1,notes6_2,notes7_1,notes8_1,notes9_1,notes9_2,notes9_3,notes9_4,notes10_1,notes11_1,hopeful_note,hopeful_note2 FROM llingjing_2")
keyword = []
for i in cursor:
    a = []
    for k in i:
        if k != "":
            a.append(k)
        else:
            pass
    keyword.append(a)
```

```python
classify2 = []
for i in classify:
    i = i.replace("生活記事","生活")
    i = i.replace("心情抒發","心情")
    i = i.replace("家庭關係","家庭")
    i = i.replace("生理狀況","生理")
    i = i.replace("反串文","反串")
    i = i.replace("就醫經驗","就醫")
    i = i.replace("、"," ")
    classify2.append(i)
```

```python
author = read_content['作者']
time = read_content['時間'].convert_objects(convert_numeric = True)
mix = list(zip(author,time))

# user = ['kid3875215 (銘哥)','angel86015 (小槑)','Astragali (Soldier A)','z881331 (絨毛雞)','s00azure (- 班長 -)','meatkai (凱)','sea212 (sea212)','DavidJam (平安又有錢才是福)','pinpo (孤獨患者)','DavidJam (平安是福)','lickmebaby (Rebirth)','ezk (蒲葵貓)','epoch3004 (木字旁)','yehz (葉子)','fervorya (一丄上止正)','viable (viable)','p1026 (豬)','gossiplarry (賴瑞)','zeze (籠中鳥)','Llingjing (冷劍塵)']

time2 = []
for k in mix:
    if k[0] == 'Llingjing (冷劍塵)':
        time2.append(k[1])
    else:
        pass
    
time3 = [] #切開日期時間空白
for n in time2:
    time3.append(n.split(" "))
```

```python
# 去除中間空白

for i in time3:
    i[1] = i[1].replace('Dec','12')
    i[1] = i[1].replace('Nov','11')
    i[1] = i[1].replace('Oct','10')
    i[1] = i[1].replace('Sep','9')
    i[1] = i[1].replace('Aug','8')
    i[1] = i[1].replace('Jul','7')
    i[1] = i[1].replace('Jun','6')
    i[1] = i[1].replace('May','5')
    i[1] = i[1].replace('Apr','4')
    i[1] = i[1].replace('Mar','3')
    i[1] = i[1].replace('Feb','2')
    i[1] = i[1].replace('Jan','1')
    if i[2] == '':
        i.remove(i[2])
    else:
        pass
```

```python
timeline = []
for i in time3:
    if i[1] == "1":
        a = "0"+i[1]
    elif i[1] == "2":
        a = "0"+i[1]
    elif i[1] == "3":
        a = "0"+i[1]
    elif i[1] == "4":
        a = "0"+i[1]
    elif i[1] == "5":
        a = "0"+i[1]
    elif i[1] == "6":
        a = "0"+i[1]
    elif i[1] == "7":
        a = "0"+i[1]
    elif i[1] == "8":
        a = "0"+i[1]
    elif i[1] == "9":
        a = "0"+i[1]
    else:
        a = i[1]
    
    if i[2] == "1":
        b = "0"+i[2]
    elif i[2] == "2":
        b = "0"+i[2]
    elif i[2] == "3":
        b = "0"+i[2]
    elif i[2] == "4":
        b = "0"+i[2]
    elif i[2] == "5":
        b = "0"+i[2]
    elif i[2] == "6":
        b = "0"+i[2]
    elif i[2] == "7":
        b = "0"+i[2]
    elif i[2] == "8":
        b = "0"+i[2]
    elif i[2] == "9":
        b = "0"+i[2]
    else:
        b = i[2]
    timeline.append(i[4]+"-"+a+"-"+b)
# timeline[::-1]
```
