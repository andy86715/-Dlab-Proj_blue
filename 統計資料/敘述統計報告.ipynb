```python
import pandas
import seaborn
import seaborn #繪圖套件
import matplotlib.pyplot as plt
data=pandas.read_excel("blue.xlsx")
```

```python

a=data["使用者ID"].value_counts()
# seaborn.countplot(x='使用者ID',data=data)
# # data['作者'].describe()
# plt.show()
# print(a)
# a.describe()
time=data["時間"].str[11:13]
# print(time)
time = time.convert_objects(convert_numeric=True)
time[time>0.5].value_counts(sort=False)
seaborn.countplot(x=time[time!=0.2],data=data)
plt.show()
sub1=data[data["標題"].str[1:3]=="陰天"]
print(len(sub1))
time=sub1["時間"].str[11:13]
time = time.convert_objects(convert_numeric=True)
time[time>0.5].value_counts(sort=False)
seaborn.countplot(x=time[time!=0.2],data=sub1)
sub2=data[data["標題"].str[1:3]=="晴天"]
print(len(sub2))
time=sub2["時間"].str[11:13]
time = time.convert_objects(convert_numeric=True)
time[time>0.5].value_counts(sort=False)
# seaborn.countplot(x=time[time!=0.2],data=sub1)
seaborn.countplot(x=time[time!=0.2],data=sub2)
plt.show()
```
