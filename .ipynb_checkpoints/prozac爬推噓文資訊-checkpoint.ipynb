```python
import requests
from bs4 import BeautifulSoup
res = requests.get("https://www.ptt.cc/bbs/prozac/index1659.html")
soup = BeautifulSoup(res.text)
for website in soup.select(".r-ent"):
    try:    
        res2 = requests.get('https://www.ptt.cc/'+website("a")[0]['href'])
        soup2 = BeautifulSoup(res2.text)
        print("--------------------------------------------------------------------------------------------------")
        for i in range(len(soup2.select(".push"))):
            print(soup2.select(".push-tag")[i].text,soup2.select(".push-userid")[i].text,":",soup2.select(".push-content")[i].text)
    except:
        pass
```
