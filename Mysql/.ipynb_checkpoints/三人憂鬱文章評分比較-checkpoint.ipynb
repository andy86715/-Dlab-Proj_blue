```python
import pymysql.cursors
import pandas as pd
```

```python
db = pymysql.connect(host="localhost", user="root", password="", db="proj_blue_server",charset='utf8mb4',port = 3310)
cursor = db.cursor()
```

```python
cursor.execute("SELECT art_newID,emotion_grade FROM emotion")
```

```python
ivy = []
for i in cursor:
    if i[1] == 0 or i[1] >=4:
        ivy.append(i)
    else:
        pass
df_ivy = pd.DataFrame(ivy)
df_ivy = df_ivy.rename(columns={0: 'ID',1: 'ivy評分'})
```

```python
cursor.execute("SELECT art_newID,emotion_grade FROM emotion_2")
```

```python
tin = []
for i in cursor:
    if i[1] == 0 or i[1] >=4:
        tin.append(i)
    else:
        pass
df_tin = pd.DataFrame(tin)
df_tin = df_tin.rename(columns={0: 'ID',1: 'tin評分'})
```

```python
cursor.execute("SELECT art_newID,emotion_grade FROM emotion_3")
```

```python
nancy = []
for i in cursor:
    if i[1] == 0 or i[1] >=4:
        nancy.append(i)
    else:
        pass
df_nancy = pd.DataFrame(nancy)
df_nancy = df_nancy.rename(columns={0: 'ID',1: 'nancy評分'})
```

```python
#ivy tin
df_ivytin = pd.merge(df_ivy,df_tin,on=['ID'])
```

```python
#ivy nancy
df_ivynancy = pd.merge(df_ivy,df_nancy,on=['ID'])
```

```python
#tin nancy
df_tinnancy = pd.merge(df_tin,df_nancy,on=['ID'])
```
